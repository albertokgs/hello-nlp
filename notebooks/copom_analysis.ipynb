```python
# Best Portuguese models for embeddings
model_name = "neuralmind/bert-base-portuguese-cased"  # or "rufimelo/bert-large-portuguese-cased-legal"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

def get_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    # Use CLS token or mean pooling
    embeddings = outputs.last_hidden_state[:, 0, :].numpy()
    return embeddings

emb1 = get_embeddings(old)
emb2 = get_embeddings(new)

similarity = cosine_similarity(emb1, emb2)[0][0]
print(f"Semantic Similarity: {similarity:.4f}")
```

```python
class CentralBankComparator:
    def __init__(self):
        self.bert_model = "neuralmind/bert-base-portuguese-cased"
        self.tokenizer = AutoTokenizer.from_pretrained(self.bert_model)
        self.model = AutoModel.from_pretrained(self.bert_model)
    
    def compare_statements(self, text1, text2):
        results = {}
        
        # 1. Semantic similarity
        emb1 = self._get_embeddings(text1)
        emb2 = self._get_embeddings(text2)
        results['similarity'] = cosine_similarity(emb1, emb2)[0][0]
        
        # 2. Length comparison
        results['length_ratio'] = len(text2) / len(text1)
        
        # 3. Key term frequency changes
        results['key_terms'] = self._compare_key_terms(text1, text2)
        
        return results
    
    def _get_embeddings(self, text):
        inputs = self.tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        with torch.no_grad():
            outputs = self.model(**inputs)
        return outputs.last_hidden_state[:, 0, :].numpy()
    
    def _compare_key_terms(self, text1, text2):
        # Economic terms common in BCB statements
        key_terms = ['inflação', 'taxa', 'juros', 'selic', 'crescimento', 
                     'pib', 'fiscal', 'crédito', 'câmbio', 'meta']
        
        comparison = {}
        for term in key_terms:
            count1 = text1.lower().count(term)
            count2 = text2.lower().count(term)
            comparison[term] = {'statement1': count1, 'statement2': count2, 'diff': count2 - count1}
        
        return comparison

# Usage
comparator = CentralBankComparator()
results = comparator.compare_statements(text1, text2)
print(results)
```
