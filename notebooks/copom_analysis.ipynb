```python
class CentralBankComparator:
    def __init__(self):
        self.bert_model = "neuralmind/bert-base-portuguese-cased"
        self.tokenizer = AutoTokenizer.from_pretrained(self.bert_model)
        self.model = AutoModel.from_pretrained(self.bert_model)
    
    def compare_statements(self, text1, text2):
        results = {}
        
        # 1. Semantic similarity
        emb1 = self._get_embeddings(text1)
        emb2 = self._get_embeddings(text2)
        results['similarity'] = cosine_similarity(emb1, emb2)[0][0]
        
        # 2. Length comparison
        results['length_ratio'] = len(text2) / len(text1)
        
        # 3. Key term frequency changes
        results['key_terms'] = self._compare_key_terms(text1, text2)
        
        return results
    
    def _get_embeddings(self, text):
        inputs = self.tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        with torch.no_grad():
            outputs = self.model(**inputs)
        return outputs.last_hidden_state[:, 0, :].numpy()
    
    def _compare_key_terms(self, text1, text2):
        # Economic terms common in BCB statements
        key_terms = ['inflação', 'taxa', 'juros', 'selic', 'crescimento', 
                     'pib', 'fiscal', 'crédito', 'câmbio', 'meta']
        
        comparison = {}
        for term in key_terms:
            count1 = text1.lower().count(term)
            count2 = text2.lower().count(term)
            comparison[term] = {'statement1': count1, 'statement2': count2, 'diff': count2 - count1}
        
        return comparison

# Usage
comparator = CentralBankComparator()
results = comparator.compare_statements(old, new)
print(results)
```

```python
# Using a local open-source model for COPOM analysis (no API key needed)
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Using a smaller multilingual model that works well with Portuguese
model_name = "Qwen/Qwen2.5-1.5B-Instruct"  # Good multilingual model, ~3GB

print("Loading model (this may take a minute on first run)...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name, 
    torch_dtype=torch.float16,
    device_map="auto"
)

def compare_with_local_llm(text1, text2):
    # Truncate texts to fit context window
    text1_short = text1[:1500]
    text2_short = text2[:1500]
    
    prompt = f"""Você é um analista econômico especializado em política monetária brasileira.

Compare as duas atas do COPOM e identifique de forma concisa:
1. Mudanças no tom (mais dovish ou hawkish)
2. Alterações nas projeções de inflação
3. Mudanças na sinalização da SELIC

ATA ANTERIOR (resumo):
{text1_short}

ATA ATUAL (resumo):
{text2_short}

Análise:"""

    messages = [{"role": "user", "content": prompt}]
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(model.device)
    
    outputs = model.generate(
        **inputs,
        max_new_tokens=500,
        temperature=0.3,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # Extract just the generated response
    return response.split("Análise:")[-1].strip()

# Usage
print("Generating analysis...")
analysis = compare_with_local_llm(old, new)
print(analysis)
```
